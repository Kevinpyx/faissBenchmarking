This document will compare the performance of a GPU and CPU for the same algorithm at 10M, 25M, and 50M. 
We will have 5 sets of plots (15 plots) in total in this document. Every set contains a plot for time per query, a plot for memory, and a plot for recall. The first three plots will compare the performance of GPU and CPU for the same algorithm at different data sizes (10M, 25M, 50M, hence three sets). The last two plots will compare the performance of the same algorithm with different data sizes (one for CPU, one for GPU, hence two sets). A hierarchical structure of the plots is shown below.

- **GPU vs CPU**: We'll compare the difference in performance of the same algorithm run on GPU and CPU (two bars per algorithm, one for gpu, one for cpu).
    - Three plots for 10M data points
        - One plot for SPQ
        - One plot for memory
        - One plot for recall
    - Three plots for 25M data points
        - One plot for SPQ
        - One plot for memory
        - One plot for recall
    - Three plots for 50M data points
        - One plot for SPQ
        - One plot for memory
        - One plot for recall
- **10M & 25M & 50M**: We'll compare the performance of the same algorithm run with different data sizes (three bars per algorithm for different sizes).
    - Three plots for CPU  
        - One plot for SPQ
        - One plot for memory
        - One plot for recall
    - Three plots for GPU
        - One plot for SPQ
        - One plot for memory
        - One plot for recall

```python
# import packages
import dataI_O as io
import numpy as np
import matplotlib.pyplot as plt
import results
```

```python
# Data preparation
def get_smr (filename): # to get the average search time, memory usage and recall. time is in ms, memory is in MB, recall is in percentage
    data = io.load_results(filename)
    avg_data = np.mean(data, axis=0)
    SPQ = avg_data[0,:,3]*1000
    memory = avg_data[0,:,4]
    recall = avg_data[0,:,5]
    return SPQ, memory, recall

def get_smr_2 (filename):
    data = io.load_results(filename)
    avg_data = np.mean(data, axis=0)
    SPQ = avg_data[:,:,3]*1000
    memory = avg_data[:,:,4]
    recall = avg_data[:,:,5]
    return SPQ, memory, recall

# FlatL2_CPU_10M
SPQ_FlatL2_CPU_10M, memory_FlatL2_CPU_10M, recall_FlatL2_CPU_10M = get_smr('/home/ypx/faissTesting/benchmark/results/FlatL2_10M_96_10k_100.h5')

# FlatL2_GPU_10M
SPQ_FlatL2_GPU_10M, memory_FlatL2_GPU_10M ,recall_FlatL2_GPU_10M = get_smr('/home/ypx/faissTesting/benchmark/results/FlatL2_10M_96_10k_100_GPU.h5')

# FlatL2_CPU_25M
SPQ_FlatL2_CPU_25M, memory_FlatL2_CPU_25M, recall_FlatL2_CPU_25M = get_smr('/home/ypx/faissTesting/benchmark/results/FlatL2_25M_96_10k_100.h5')

# FlatL2_GPU_25M
SPQ_FlatL2_GPU_25M, memory_FlatL2_GPU_25M ,recall_FlatL2_GPU_25M = get_smr('/home/ypx/faissTesting/benchmark/results/FlatL2_25M_96_10k_100_GPU.h5')

# FlatL2_CPU_50M
SPQ_FlatL2_CPU_50M, memory_FlatL2_CPU_50M, recall_FlatL2_CPU_50M = get_smr('/home/ypx/faissTesting/benchmark/results/FlatL2_50M_96_10k_100.h5')

# FlatL2_GPU_50M
SPQ_FlatL2_GPU_50M, memory_FlatL2_GPU_50M, recall_FlatL2_GPU_50M = get_smr('/home/ypx/faissTesting/benchmark/results/FlatL2_50M_96_10k_100_GPU.h5')

# LSH_CPU_10M
SPQ_LSH_CPU_10M, memory_LSH_CPU_10M, recall_LSH_CPU_10M = get_smr('/home/ypx/faissTesting/benchmark/results/LSH_10M_96_10k_100.h5')

# LSH_GPU_10M
SPQ_LSH_GPU_10M, memory_LSH_GPU_10M, recall_LSH_GPU_10M = get_smr('/home/ypx/faissTesting/benchmark/results/LSH_10M_96_10k_100_GPU.h5')

# LSH_CPU_25M
SPQ_LSH_CPU_25M, memory_LSH_CPU_25M, recall_LSH_CPU_25M = get_smr('/home/ypx/faissTesting/benchmark/results/LSH_25M_96_10k_100.h5')

# LSH_GPU_25M
SPQ_LSH_GPU_25M, memory_LSH_GPU_25M, recall_LSH_GPU_25M = get_smr('/home/ypx/faissTesting/benchmark/results/LSH_25M_96_10k_100_GPU.h5')

# LSH_CPU_50M
SPQ_LSH_CPU_50M, memory_LSH_CPU_50M, recall_LSH_CPU_50M = get_smr('/home/ypx/faissTesting/benchmark/results/LSH_50M_96_10k_100.h5')

# LSH_GPU_50M
SPQ_LSH_GPU_50M, memory_LSH_GPU_50M, recall_LSH_GPU_50M = get_smr('/home/ypx/faissTesting/benchmark/results/LSH_50M_96_10k_100_GPU.h5')

# PQ_CPU_10M
SPQ_PQ_CPU_10M, memory_PQ_CPU_10M, recall_PQ_CPU_10M = get_smr_2('/home/ypx/faissTesting/benchmark/results/PQ_10M_96_10k_100.h5')

# PQ_GPU_10M
SPQ_PQ_GPU_10M, memory_PQ_GPU_10M, recall_PQ_GPU_10M = get_smr_2('/home/ypx/faissTesting/benchmark/results/PQ_10M_96_10k_100_GPU.h5')

# PQ_CPU_25M
SPQ_PQ_CPU_25M, memory_PQ_CPU_25M, recall_PQ_CPU_25M = get_smr_2('/home/ypx/faissTesting/benchmark/results/PQ_25M_96_10k_100.h5')

# PQ_GPU_25M
SPQ_PQ_GPU_25M, memory_PQ_GPU_25M, recall_PQ_GPU_25M = get_smr_2('/home/ypx/faissTesting/benchmark/results/PQ_25M_96_10k_100_GPU.h5')

# PQ_CPU_50M
SPQ_PQ_CPU_50M, memory_PQ_CPU_50M, recall_PQ_CPU_50M = get_smr_2('/home/ypx/faissTesting/benchmark/results/PQ_50M_96_10k_100.h5')

# PQ_GPU_50M
SPQ_PQ_GPU_50M, memory_PQ_GPU_50M, recall_PQ_GPU_50M = get_smr_2('/home/ypx/faissTesting/benchmark/results/PQ_50M_96_10k_100_GPU.h5')

# HNSW_CPU_10M
SPQ_HNSW_CPU_10M, memory_HNSW_CPU_10M, recall_HNSW_CPU_10M = get_smr('/home/ypx/faissTesting/benchmark/results/HNSWFlat_10M_96_10k_100.h5')

# HNSW_CPU_10M_4
SPQ_HNSW_CPU_10M_4, memory_HNSW_CPU_10M_4, recall_HNSW_CPU_10M_4 = get_smr('/home/ypx/faissTesting/benchmark/results/HNSWFlat_10M_96_10k_100_4.h5')

# HNSW_GPU_10M
SPQ_HNSW_GPU_10M, memory_HNSW_GPU_10M, recall_HNSW_GPU_10M = get_smr('/home/ypx/faissTesting/benchmark/results/HNSWFlat_10M_96_10k_100_GPU.h5')

# HNSW_GPU_10M_4
SPQ_HNSW_GPU_10M_4, memory_HNSW_GPU_10M_4, recall_HNSW_GPU_10M_4 = get_smr('/home/ypx/faissTesting/benchmark/results/HNSWFlat_10M_96_10k_100_GPU_4.h5')

# HNSW_CPU_25M
SPQ_HNSW_CPU_25M, memory_HNSW_CPU_25M, recall_HNSW_CPU_25M = get_smr('/home/ypx/faissTesting/benchmark/results/HNSWFlat_25M_96_10k_100.h5')

# HNSW_GPU_25M
SPQ_HNSW_GPU_25M, memory_HNSW_GPU_25M, recall_HNSW_GPU_25M = get_smr('/home/ypx/faissTesting/benchmark/results/HNSWFlat_25M_96_10k_100_GPU.h5')

# HNSW_CPU_50M
SPQ_HNSW_CPU_50M, memory_HNSW_CPU_50M, recall_HNSW_CPU_50M = get_smr('/home/ypx/faissTesting/benchmark/results/HNSWFlat_50M_96_10k_100.h5')

# HNSW_GPU_50M
SPQ_HNSW_GPU_50M, memory_HNSW_GPU_50M, recall_HNSW_GPU_50M = get_smr('/home/ypx/faissTesting/benchmark/results/HNSWFlat_50M_96_10k_100_GPU.h5')


# parameter categories
# LSH
cat_LSH_nbits_10M = [2, 6, 8, 16, 24, 32, 64, 128, 256, 512, 768, 1024]
cat_LSH_nbits_25M = [2, 6, 8, 16, 24, 32, 64, 128, 256]
cat_LSH_nbits_50M = [2, 6, 8, 16, 24, 32, 64, 128]

# PQ
cat_PQ_m_10M = [8, 32]
cat_PQ_nbits_10M = [8, 10]
cat_PQ_m_25M = [8, 32]
cat_PQ_nbits_25M = [8, 10]
cat_PQ_m_50M = [8, 32]
cat_PQ_nbits_50M = [8, 10]

# HNSW
cat_HNSW_M_10M_4 = [4]
cat_HNSW_M_10M = [16, 32, 64, 96]
cat_HNSW_M_25M = [4, 8]
cat_HNSW_M_50M = [4]
```

For LSH, the GPU implementation runs slower than the CPU implementattion for all data sizes tested. 
For PQ, the GPU implementation runs faster than the CPU implementation for some parameter settings, but slower for others. To be more specific, when the data size is 10M or 25M, the GPU implementation runs faster than the CPU implementation when m = 32, and the other way around when m = 8. When the data size is 50M, the GPU implementation runs faster than the CPU implementation under all parameter settings tested. We observe that while increasing m and nbits will both pronounce the difference between CPU and GPU implementations, increasing m has a more significant impact.
For HNSW, looking at the chart with data size = 10M along, we observe that increasing M will pronounce the difference between CPU and GPU implementations. However, when data size is 25M or 50M, due to lack of memory, we are not able to test the impact of M.

Consequently, when there is a memory constraint, both PQ and LSH would be superior to the exact search algorithm. Those prioritizing search efficiency may favor LSH, while PQ excels in achieving higher quality results albeit with extended computation time. In contrast, hierarchical navigable small world entails greater memory overhead than even the exact search due to its dependency on constructing a comprehensive vector graph for searches. However, the graph structure enables it to perform the search one-thousand times faster than other algorithms with a fair recall (0.7-0.8). Therefore, when ample memory resources and rapid search times are of significance, HNSW stands as an evident choice.
