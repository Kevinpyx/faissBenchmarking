This document will compare the performance of a GPU and CPU for the same algorithm at 10M, 25M, and 50M. 
We will have 5 sets of plots (15 plots) in total in this document. Every set contains a plot for time per query, a plot for memory, and a plot for recall. The last two plots will compare the performance of the same algorithm with different data sizes (one for CPU, one for GPU, hence two sets). A hierarchical structure of the plots is shown below.
